### Data Cleaning (We identified a total of three columns needs to be cleaned further)

>Create three columns - 
1. representing the day with in the week for Scheduled and appointed days
2. Represeting the delta between these two dates

```python
# After discussing the structure of the data and any problems that need to be
# cleaned, perform those cleaning steps in the second part of this section.
#Weekday and month is extracted from all date columns. We would get ScheculedDayWeekday & appointment Day weekday column represet day of week
#Month will be in form of MM-YYYY format
#To understand the format of strftime, I referred this website - https://www.journaldev.com/23365/python-string-to-datetime-strptime

date_columns = ['scheduledday', 'appointmentday']
for c in date_columns:
    df[c] = pd.to_datetime(df[c])
    df[c] = df[c].dt.date
    c_temp=c+'weekday'
    df[c_temp] = df[c].apply(lambda x: dt.datetime.strftime(x, '%w'))
    df[c_temp] = pd.to_numeric(df[c_temp])
    c_month = c+'month'
    df[c_month] = df[c].apply(lambda x: dt.datetime.strftime(x, '%m-%Y'))
    df[c_month] = pd.to_datetime(df[c_month])
        
df.info()
```


>Fixing the age column

```python
#Correcting age columns

df.loc[df['age'] < 0, 'age'] = 0
df.head()

bins = [0,18,30,50,70,120]
labels=['0-18','19-30','31-50','51-70','71-120']
df['bins'] = pd.cut(df['age'], bins=bins, labels=labels, include_lowest=True)
df.head()
```

```python
#One hot encoding the categorical columns

df = pd.get_dummies(df, columns=['gender','no_show'], drop_first=False)

df.head()
    
```

### Understand the current picture of no-shows - Are there any geographies, where there is difference in No-Show appointments?

>Grouping the geographies on the basis of No-shows
1. First, we will find the ratio of no shows and appointments by region
2. After that, we will box plot these values to understand the distribution pattern 
3. Up on that, we will decide on definition of Good, Medium, Bad Adherence further

```python
no_show_pct = df.groupby('neighbourhood').no_show_Yes.apply(lambda x: x.sum()/x.count())

no_show_pct = pd.DataFrame(no_show_pct)

no_show_pct.columns
```

```python
#http://cmdlinetips.com/2018/03/how-to-make-boxplots-in-python-with-pandas-and-seaborn/ - Box plot visualization credits
#Using Pandas - no_show_pct[(no_show_pct['no_show_Yes']<1) & (no_show_pct['no_show_Yes']>0)].boxplot(column=['no_show_Yes'],grid=False)

bplot = sns.boxplot(y='no_show_Yes', data=no_show_pct[(no_show_pct['no_show_Yes']<1) & (no_show_pct['no_show_Yes']>0)], 
                 width=0.25,
                 palette="colorblind")
```

```python
no_show_pct['Adherence_Group']=pd.qcut(no_show_pct['no_show_Yes'],3, labels=["1_good", "2_medium", "3_bad"])
no_show_pct.groupby('Adherence_Group').no_show_Yes.max()
```

>The plot here represnets 'No-show %' distribution by various geographies. THe values range from as low as 5% to 30% (Note - We eliminated data points at two extremes as these case exists because of the fewer appointments generated from these places). On the basis of the distribution, we will split the Good, Medium and Bad with following criteris -
1. Good - <=33rd Percentile (<18%) of No show % 
2. Medium - Between 34th and 66th Percentile of No Show% (18% to 22%)
3. Bad - Above 66th Percentile range (>22%)

```python
#Append this information to the main dataset

df_append = pd.merge(df,no_show_pct,how='inner',left_on='neighbourhood',right_index=True)

df_append.head()
```

### How are the patient characteristics different between the patients from the three Adherence groups?

>We will define a fucntion, that computes two summary measures - 
1. It will first test if the two groups of bi-variate are statistically significant (Please note that all the variables in the data set are categorical. As a result we only considered chi-square test for categorical variables here)
2. Get the distribution of patient appointments acorss the adherence group
3. Get the distribution of patient no-shows/Ratio of No-shows to Appointmens across the adherence group
4. Plot them in a column/bar chart


```python
def bivariate_func(cols):
    
    #Significance testing to check if the analysis that we are going to do is statistically significant
    crosstab = pd.crosstab(df_append['Adherence_Group'], df_append[cols])
    
    chi2_stat, p_val, dof, ex = stats.chi2_contingency(crosstab)
    
    if p_val<0.05:
        string = "P-value is below the 0.05 CI. We can reject the null hypothesis"
    else:
        string = "P-value is not significant"
        
    print(string)
    #Find no show % for the selected Variables list
    
    No_Show_Pct=df_append.groupby(['Adherence_Group',cols])['no_show_Yes_x'].apply(lambda x:(x.sum()/x.count())).reset_index()
       
    #Find %Patients by the defined column with in Column given
    
    def agg_fun(a,b):
        Pct1 = pd.DataFrame(df_append.groupby([a,b]).no_show_Yes_x.count()/df_append.groupby([a]).no_show_Yes_x.count())
        Pct2 = pd.DataFrame(df_append.groupby([a,b]).no_show_Yes_x.sum()/df_append.groupby([a]).no_show_Yes_x.sum())
        Pct1.reset_index(inplace=True)
        Pct2.reset_index(inplace=True)
        final_frame = pd.merge(left = Pct1,right=Pct2,left_on=[a,b],right_on=[a,b],suffixes=('_PPL', '_PCT')) 
        return final_frame

    Pct_Population=agg_fun('Adherence_Group',cols)
    
    #For output representation
    
    final_frame = pd.merge(left = Pct_Population,right=No_Show_Pct,left_on=['Adherence_Group',cols],right_on=['Adherence_Group',cols])

    final_frame.rename(columns={'no_show_Yes_x_PPL':'Appointments_in_Group','no_show_Yes_x_PCT':'Noshows_in_Group','no_show_Yes_x':'Noshows_pct'},inplace=True)
    
    print(final_frame)
    #print(cols+'Plot to understand No-Show rate')
    df_append.groupby([cols])['no_show_Yes_x'].apply(lambda x:(x.sum()/x.count())).plot.bar()
    plt.title(cols+'Plot to understand No-Show rate')
    plt.xlabel(cols)
    plt.ylabel('% Noshows out of Appoitments')
    
    def chart_updater(df_temp,xbar,ybar,var,title,stack):
        df_temp.sort_values(by= ybar)
        my_list = df_temp[ybar].tolist()
        my_list = list(set(my_list))
        my_list.sort()
        pivot_df = df_temp.pivot(index=xbar, columns=ybar, values=var)     
        pivot_df.loc[:,my_list].plot.bar(stacked=stack, figsize=(10,7), fontsize=13)
        plt.title(title)
        plt.xlabel(xbar)
        plt.ylabel(ybar);

    chart_updater(final_frame,'Adherence_Group',cols,'Appointments_in_Group','% of Appointments in Adherence Group',False)
    chart_updater(final_frame,'Adherence_Group',cols,'Noshows_in_Group','% of Noshows with in Group',False) 
    chart_updater(final_frame,'Adherence_Group',cols,'Noshows_pct','Ratio of No-shows to appointments',False) 
    
    return final_frame
```

### Gender dynamics

>Gender Analysis - Men in _Bad_ adherence regions exhibit higher intake of appointments as well as higher no-show rate compared to the men in _Good_ & _Medium_ adherence regions
